### 2 - Manipulando Dados em Python: Números, Textos e Variáveis

Manipular dados é fundamental para qualquer projeto de programação, desde *scripts* simples até análise de dados complexa. A ling. Python facilita esse processo com uma sintaxe clara e tipos de dados dinâmicos, onde a manipulação se baseia em números, textos (strings) e variáveis.

#### Variáveis em Python
As variáveis são armazenadas localmente na memória RAM do computador.  

* **Tipagem Dinâmica**: Não é preciso declarar o tipo da variável antes de usá-la, assim como ocorre em outras linguagens; o interpretador Python a define automaticamente.
* **Declaração**: ` idade = 25, nome = "Ana"`.
* **Reatribuição**: O valor de uma variável pode mudar seu valor e seu tipo durante a execução do código.
* **Função `type()`**: Retorna o tipo de dado de uma variável.

````python
idade = 50
nome = "Ana"
print("Nome: ", nome,"\nIdade: ", idade)
print(type(nome), type(idade))
````

#### Manipulação de Números
Python gerencia dois tipos principais de números:
* **Inteiros (int)**: Números inteiros não possuem casa(s) decimal(is); `10, -5`
* **Ponto Flutuante (float)**: Números com parte fracionária; `-10.5, 3.1415926535`
* **Operações Aritméticas**: Soma (+), Subtração (-), Multiplicação (*), Divisão (/), Divisão Inteira (//), Resto da divisão (%) e Potenciação (**)
* **Formatação**: Podemos usar o formato `.2f` para limitar em duas casas decimais a exibição do valor de uma variável do tipo `float`

As `F-strings` oferecem a sintaxe mais legível para formatações: `f"{variavel:formato}"`.   
Permitem definir casas decimais (`.2f`), alinhar texto, adicionar símbolos de moeda (R$) ou porcentagem (%) e formatar datas.

````python
print(f"{-10.5:.5f}\nPi ={3.1415926535:10.3f}")
````

#### Manipulação de Textos (*Strings*)
*Strings* são sequências de caracteres delimitadas por aspas simples (' ') ou duplas (" "). 
* **Concatenação**: Une *strings* usando o operador '+'; `nome + " Silva"`
* **Acesso** e **Fatiamento (*Slicing*)**: Acessar partes do texto: `texto[0]` pega a primeira letra, `texto[2:5]` pega da terceira à quinta letra
* **Métodos Importantes**
    `len()`: Conta o número de caracteres.  
    `strip()`: Remove espaços em branco antes e depois da string.  
    `upper()` / `.lower()`: Converte para maiúsculas ou minúsculas.  
    `split()`: Divide a string em uma lista de strings.  
    `find()`: Localiza uma substring dentro do texto.  

````python
nome = "João"
sobrenome = "Silva"
print(nome+sobrenome)
print(nome+" "+sobrenome)
print("Sigla:", nome[0]+sobrenome[0])
print(nome.upper(), sobrenome.upper())
print("Posição da substring 'il':", sobrenome.find('il'))
````

#### Conversão de Tipos (*Type Casting*)
É possível converter dados para formatos diferentes para garantir que a operação funcione corretamente: 
* `int()  `: Converte para inteiro; `int("10") --> 10`
* `float()`: Converte para número decimal; `float("10.5") --> 10.5`
* `str()  `: Converte números para texto;  `str(-25.1) --> "-25.1"`  

````python
nome = "Python"
versao = 3.12
# Usando f-string para formatar texto e números:
mensagem = f"Bem-vindo ao {nome} na versão {versao}"
print(mensagem.upper()) # Saída: BEM-VINDO AO PYTHON NA VERSÃO 3.1
````

### 3 - Estruturas de Dados Nativas: Listas [ ] e Tuplas ( )

Tratam-se de estruturas de dados nativas para armazenagem de dados ordenados em Python.  
As **Listas** são mutáveis (alterações, acréscimos ou remoções de itens são permitidos), ideais para coleções dinâmicas de dados.  
As **Tuplas** são imutáveis (não podem ser alteradas após a criação), sendo mais rápidas, seguras para dados constantes e ideais para registros.

#### Diferenças e Características:
* **Listas** (list):

    Sintaxe: `minha_lista = [1, 2, "python", 3.14]`  
    Operações: `.append()` para anexação de itens, `.remove()` para retirada de itens, ou alterações via uso de índice do item: `lista[0] = 10`  
    Uso: Coleções de dados heterogêneos que mudam de tamanho ou valor frequentemente

* **Tuplas** (tuple):

    Sintaxe: `minha_tupla = (1, 2, "python", 3.14)`  
    Operações: por serem Imutáveis, uma vez criadas, não mudam  
    Uso: Dados estruturais, constantes, retornos de funções (mais rápido e seguro)  

* **Comportamentos** comuns a ambas:  
    Ordenadas: Mantêm a ordem de inserção  
    Indexação: Acessíveis por [0], [-1], fatiamento [1:3]  
    Heterogêneas: Podem conter tipos de dados diferentes (inteiros, strings etc.)

    Use **listas** quando precisar alterar o conteúdo ao longo do tempo. Use **tuplas** para garantir a integridade dos dados e otimizar o desempenho quando a estrutura não precisar de alterações. 

````python
lista_teste = ['Engenheiro', 'Médico', 'Arquiteto']
print(type(lista_teste), lista_teste)
print(lista_teste[0])
print(lista_teste[1])
print(lista_teste[2])
````

````python
lista_teste = ['Engenheiro', 'Médico', 'Arquiteto', 'Engenheiro']
print(type(lista_teste), lista_teste)
print(lista_teste[0])
print(lista_teste[1])
print(lista_teste[2])
print(lista_teste[3])
lista_teste[3] = 'Farmacêutico'
print(lista_teste)
````

````python
outra_lista = ['Engenheiro', 10, 10.5, 'Médico', -1]
print(outra_lista)
print(outra_lista[0])
print(outra_lista[3])
print(outra_lista[4])
print(outra_lista[0:4])
outra_lista.append(1001)
print(outra_lista)
````

````python
tupla1 = (23, "ok")
tupla2 = tuple(outra_lista)
print(tupla1, tupla2)
print(tupla1[0])
# tupla1[0] = 25
tupla1.append(1001)
````

#### Pilhas e Filas
Python não possui estruturas nativas para **Pilhas** ou **Filas**, embora possamos implementá-las usando **listas**.  

**Pilha**  
Pilha é uma lista de elementos onde novos elementos são adicionados somente no final (topo) da lista, ou removidos somente do final (topo) da lista. 
Numa pilha de livros, sempre que adicionamos livro(s) a pilha cresce, e sempre que removemos livro(s) a pilha diminui. Essa estrutura de dados é baseado no princípio **"último a entrar, é o primeiro a sair"** (LIFO).

````python
pilha = [1, 2, 3, 4]
print(pilha)
pilha.append(5)       # acrescenta item no topo da pilha
print(pilha)
pilha.pop()           # retira item do topo da pilha
print(pilha)
````

**Filas**  
Fila é uma lista de elementos na qual novos elementos são adicionados ao final da lista, e os elementos são removidos do início da lista. É comparável a ficar numa fila no mundo real. A fila fica mais curta à medida que os que estão na frente são atendidos. Fica mais longa quando uma nova pessoa entra no final da fila. Essa estrutura de dados é baseado no princípio **"primeiro a entrar, é o primeiro a sair"** (FIFO).

````python
fila = [1, 2, 3, 4]
print(fila)
fila.append(5)        # acrescenta item no final da fila
print(fila)
fila.pop(0)           # retira item do início da fila
print(fila)
````

### 4 - Controlando o Fluxo de Execução: if / elif / else 

O controle de fluxo com `if, elif` e `else` em Python permite que o programa tome decisões, executando blocos de código específicos com base em condições booleanas (`True/False`). O `if` avalia uma condição, `elif`  (abreviação de `else if`) verifica condições adicionais e `else` define o comportamento se nenhuma condição anterior for verdadeira.

#### Sintaxe
* **Identação**: Python usa identação (espaços no início da linha) compulsoriamente para definir o bloco de código de cada condição.
* **Dois Pontos (:)**: As instruções `if, elif` e `else` devem terminar com dois pontos.
* **Operadores Relacionais**: Utilizados para comparar valores, `==` (igualdade), `!=` (diferente), `>`, `<`, `>=`, `<=`.
* **Operadores Lógicos**: Combinam expressões relacionais, `and` (e), `or` (ou), `not` (não).
* **If Curto** (Ternário): É uma estrutura condicional simples escrita numa única linha: `resultado = "Aprovado" if nota >= 7 else "Reprovado"`

````python
# Exemplo de uso da estrutura condicional de fluxo de execução
nota = 7.5
if nota >= 9:
    print("Nota A")
elif nota >= 7:
    print("Nota B")
else:
    print("Nota C")
````

````python
idade = 17
if idade < 13:
    print("Criança")
elif 13 <= idade < 18:
    print("Adolescente")
else:
    print("Adulto")
````

### 6 - Controlando o Fluxo de Execução: while / for

O controle de fluxo em Python utiliza `for` para iterar sobre sequências (listas, tuplas, strings, range) um número definido de vezes, e `while` para repetir códigos enquanto uma condição for verdadeira (ideal para números de iterações desconhecidos). Ambos laços podem ser interrompidos por `break` e `continue` (força a próxima iteração).

#### Laço `for` (Iteração Definida)
O `for` em Python é usado para percorrer os itens de uma sequência na ordem em que aparecem.

````python
# Iterando sobre uma lista
salada = ["maçã", "banana", "uva"]
for fruta in salada:
    print(fruta)
````

````python
# Usando a função nativa range() para realizar 5 repetições (0 a 3)
for i in range(4):
    print(f"Passo {i}")
````

#### Laço `while` (Repetição Condicional)
O `while` repete um bloco de código enquanto a condição especificada for `True`. É crucial garantir que a condição se torne `False` eventualmente para evitar laços infinitos.

````python
# Contador de passos
contador = 0
while contador < 4:
    print(contador)
    contador += 1      # Essencial para evitar laço infinito
````

#### Interrupção e Continuação de Laços
* `break`: Finaliza o laço imediatamente.
* `continue`: Ignora o restante do código na iteração atual e pula para a próxima iteração.

````python
# break
for n in range(10):
    if n >= 4:
        break         # Para o loop quando n for 4
    print(n)
````

````python
# continue
for n in range(5):
    if n == 2:
        continue      # Pula a impressão do número 2
    print(n)
````

### 7 - Mais Estruturas de Dados: Dicionários

São estruturas de dados nativas, mutáveis e que armazenam pares ordenados de `chave:valor`. Eles funcionam como tabelas *hash* usadas em busca rápida, permitindo acesso, inserção e exclusão eficientes. **`Chaves`** devem ser constantes únicas e imutáveis (strings, números, tuplas), enquanto **`valores`** podem ser de qualquer tipo.

Dicionários são também chamados de “memória associativa” ou “vetor associativo” em outras linguagens. Diferente das sequências que são indexadas por valores numéricos inteiros, os dicionários são indexados por `chaves` (*keys*), que podem ser de qualquer tipo imutável (como strings e inteiros).

#### Características
* Mutabilidade: Os dicionários podem ser alterados, adicionando, removendo ou modificando elementos após a sua criação.
* Mapeamento: Acesso a valores através de chaves únicas, índices não numéricos.
* Performance: A busca de um item é extremamente rápida, ideal para armazenar grandes coleções de dados estruturados.
* Criação: Utiliza-se um par de chaves `{}` e dois pontos para separar `chaves` e `valores`; `meu_dict = {"nome": "Ana", "idade": 25}`
* Métodos Comuns:

    `dic = {}`: cria um dicionário vazio.
    `dic[chave] = valor`: Adiciona ou atualiza um item no dicionário.  
    `del dic[chave]`: Remove o item descrito pela 'chave' do dicionário.  
    `dic.keys()`: Retorna todas as chaves do dicionário.  
    `dic.values()`: Retorna todos os valores do dicionário.  
    `dic.items()`: Retorna todos os pares (chave, valor) do dicionário.  

````python
# Cria um dicionário
usuario = {"id": 1, "nome": "João", "função": "admin"}
# Acessa um valor
print(usuario["nome"])
# Adiciona/Modifica item(ns)
usuario["ativo"] = True
usuario["id"] = 2
# Verifica
if "função" in usuario:
    print("Função definida!")
````

### 8 - Modularizando o Código: Funções do Usuário

Modularizar código é o processo de dividir um programa grande e complexo em partes menores, organizadas e reutilizáveis, conhecidas como **módulos**. Isso facilita a manutenção, teste e leitura do código, permitindo que alterações sejam feitas em uma parte do código sem afetar o todo.

````python
def area_retangulo(base, altura):
    """ Calcula a área de um retângulo """
    area = base * altura
    return area

# Chamando a função
resultado = area_retangulo(10, 5)
print(rf"A área é: {resultado} m2")
````

### 9 - Escrevendo o Código com Classes

Escrever código com **Classes** em Python é a base da **Programação Orientada a Objetos** (POO). Ela permite organizar o código agrupando dados (atributos) e comportamentos (métodos) em unidades lógicas. 

#### Estrutura de uma Classe
A estrutura envolve o uso da palavra-chave `class`, seguida pelo nome da classe (em geral se usa PascalCase - convenção de nomenclatura, na qual a primeira letra de cada palavra num nome composto é maiúscula, sem espaços ou caracteres especiais. Ex.: `MinhaClasseExemplo`) e o método **construtor** `__init__`, que define os atributos iniciais do objeto.

**Exemplo**: Classe de Carros

````python
class Carro:
    
    def __init__(self, marca, modelo, ano):                            # Método construtor (inicializa os atributos)
        self.marca = marca      # Atributo
        self.modelo = modelo    # Atributo
        self.ano = ano          # Atributo
        self.velocidade = 0     # Atributo com valor padrão (default)

    def acelerar(self):                                                # Método (comportamento)
        self.velocidade += 10
        print(f"O {self.modelo} acelerou para {self.velocidade} km/h")

    def exibir_info(self):                                             # Método para exibir informações
        print(f"Carro: {self.marca} {self.modelo}, Ano: {self.ano}")
````

#### Instanciando Objetos da Classe "Carro"
Criar um objeto a partir da classe é chamado de instanciação. A classe é vista como um esqueleto virtual, a partir da qual a instanciação cria uma entidade na memória a partir do esqueleto descrito pela classe, ou seja, um objeto. 

````python
# Criando objetos (instâncias da classe Carro)
carro1 = Carro("Fiat", "Argo", 2025)
carro2 = Carro("Ford", "Territory", 2026)

# Acessando métodos e atributos
carro1.exibir_info()  # mostra os dados do objeto 'carro1"
carro1.acelerar()     # acelera o objeto 'carro1" para a veloc. 10 km/h
carro2.exibir_info()  # mostra os dados do objeto 'carro2"
````

#### Conceitos da POO em Python
* `self`: Representa a própria instância do objeto que está sendo manipulado. Ele é usado para acessar os atributos e métodos da classe.
* `__init__`: Método especial, **construtor** do objeto, chamado automaticamente quando um novo objeto é criado, servindo para inicializar os atributos.
* Métodos: Funções definidas na classe, e que descrevem os comportamentos do objeto.

**Exemplo**
Conta Bancária: Classe para gerenciar estados, tal como **saldo da conta bancária**.

````python
class ContaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self.saldo = saldo_inicial

    def depositar(self, valor):
        self.saldo += valor
        print(f"Depósito de R$ {valor:.2f} realizado. Saldo atual: R$ {self.saldo:8.2f}")

    def sacar(self, valor):
        if valor <= self.saldo:
            self.saldo -= valor
            print(f"Saque de R$ {valor:.2f} realizado.    Saldo atual: R$ {self.saldo:8.2f}")
        else:
            print("Saldo insuficiente!")

# Uso
minha_conta = ContaBancaria("João", 1000)
minha_conta.depositar(500)
minha_conta.sacar(700)
````

#### Boas Práticas
* **Documentação**: Utilize `docstrings` para explicar a classe e seus métodos.
* **Nomenclatura**: Nomes de classe em PascalCase (ex: MinhaClasse), métodos e variáveis em *snake_case* (ex: nome_variavel).
* **Método __str__**: Defina o método `__str__` para retornar uma string legível sobre o objeto, útil para o comando `print()`.

**Docstrings em Python**: são strings literais usadas para documentar módulos, funções, classes ou métodos, sendo colocadas logo abaixo da definição, utilizando três aspas duplas ("""). Diferente de comentários, elas persistem em tempo de execução, permitindo o uso da função `help()` e ferramentas automáticas para gerar documentação técnica.

### 10 - Manipulando Arquivos de Dados e Erros

A manipulação de arquivos e o tratamento de erros são aspectos fundamentais na programação em Python, permitindo que *scripts* leiam, escrevam e gerenciem dados de forma segura.  

#### Manipulação de Arquivos de Dados
A forma mais comum de trabalhar com arquivos de dados é usando a função `open()`, que abre um arquivo e retorna um objeto de arquivo. É altamente recomendável usar a instrução `with`, pois ela garante que o arquivo seja fechado automaticamente após o seu uso. 

#### Principais Modos de Abertura de um Arquivo de Dados:
* `'r'` (Ler): Abre para leitura (padrão). O arquivo deve existir, caso contrário um erro será indicado.
* `'w'` (Escrever): Abre para escrita/gravação. Cria o arquivo se ele não existir, ou apaga o conteúdo se ele já existir.
* `'a'` (Anexar): Abre para adicionar dados ao final do arquivo, sem apagar o conteúdo existente.
* `'b'` (Binário): Usado para arquivos não-texto (imagens, áudios, vídeos).

Os modos de abertura podem ser combinados, tais como: `'rb'` para abrir arquivo bináro para leitura, ou `'wb'` para abrir arquivo binário para gravação.

#### Exemplo de Manipulação de Arquivo de : modo "iniciante"

````python
# Exemplo abertura de arquivo de dados no modo de Escrita
arquivo = open('exemplo.txt', 'w', encoding='utf-8')
arquivo.write('Oi pessoal!')
arquivo.close()

# Exemplo de abertura de arquivo de dados no modo de Leitura
arquivo = open('exemplo.txt', 'r', encoding='utf-8')
conteudo = arquivo.read()
print(conteudo)
arquivo.close()
````

#### Exemplo de Manipulação de Arquivo de Dados: modo "avançado"
Usando o comando `with` que faz o fechamento automático do arquivo depois da manipulação (escrita ou leitura) dos dados do arquivo.

````python
# Exemplo abertura de arquivo de dados no modo de Escrita
with open('novo_exemplo.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write('Oi pessoal, sejam bem-vindos!')

# Exemplo de abertura de arquivo de dados no modo de Leitura
with open('novo_exemplo.txt', 'r', encoding='utf-8') as arquivo:
    conteudo = arquivo.read()
    print(conteudo)
````

#### Tratamento de Erros e Exceções (*try/except*) 
Para evitar que um *script* pare inesperadamente, podemos utilizar um bloco `try-except` para capturar possíveis exceções, como por exemplo, tentar abrir um arquivo que não existe (`FileNotFoundError`).  

**Exemplo**: Finalizando um *script* de forma elegante, sem apresentação de mensagens do interpretador.

````python
try:
    with open('arquivo_inexistente.txt', 'r') as arquivo:
        conteudo = arquivo.read()
except FileNotFoundError:
    print("Erro: O arquivo não foi encontrado!")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
finally:
    print("Execução finalizada.")
````

### 11 - Teste de Unidade

O **teste de unidade** é um método de teste de software pelo qual unidades individuais de código-fonte são submetidas a vários testes para determinar se estão aptas para o uso. Ele verifica a qualidade do código testado.

Ao concluir o processo de desenvolvimento do código, o desenvolvedor codifica os resultados que são conhecidos como potencialmente práticos e úteis no *script* de teste para verificar a correção de uma **unidade** (módulo, função, sub-rotina, trecho etc.) específica. Durante a execução do **caso de teste**, várias estruturas registram os testes que falham em qualquer critério e os relatam num resumo.

Os desenvolvedores escrevem *scripts* de teste automatizados, garantindo que cada **seção** ou **unidade** do código atenda ao projeto e se comporte conforme o esperado.  

A estrutura (pacote) de **teste de unidade** em Python é chamada **unittest**, e trata-se de um pacote nativo da Ling. Python. 

#### Tipos de Unidades

* Um módulo inteiro
* Uma função individual
* Uma interface completa (uma classe ou um método)
A melhor maneira de escrever testes unitários para o código é começar com a menor unidade testável que o seu código possa ter, depois passar para outras unidades maiores e ver como a menor unidade testada interage com as outras unidades, assim criamos um **teste de unidade** abrangente para os aplicativos.

Os testes de unidades do Python permite:
* Automação de testes
* Compartilhamento do código de configuração e desligamento para testes
* Agregação de testes em coleções
* Independência dos testes em relação à estrutura de relatórios

#### Usando o Pacote **unittest**

Vamos, a título de exemplo, criar uma função para calcular o volume de um cubo.

````python
def cubo_volume(lado):
    return (lado*lado*lado)

lados = [2, 1.1, -2.5, 2j, 'two']
for i in lados:
    print("Volume do cubo:", cubo_volume(i))
````

#### Resultados
Observamos alguns erros (três) na execução da unidade (função):
1. O volume do cubo não pode ser negativo,
1. O volume do cubo não deve ser um número complexo,
1. O código resulta em um `TypeError`, pois não se pode multiplicar uma cadeia de caracteres (string).  

Os testes de unidade geralmente são escritos como um código separado em um arquivo diferente, e pode haver diferentes convenções de nomenclatura a seguir. Podemos escrever o nome do arquivo de teste de unidade como "nome_do_arquivo_" + "test" + ".py".

Para usarmos o pacote **unittest** precisamos preparar o ambiente, como mostrado a seguir:

#### Preparação (Código e Testes na mesma célula)
O cenário mais comum é definir a função (unidade) e criar o teste na mesma célula ou em células subsequentes.

**Exemplo**: Digamos que o nome do arquivo do código seja `cubo_volume.py`, então o nome do arrquivo com o código de teste de unidade poderia ser `cubo_volume_test.py`.

````python
import unittest
# from cubo_volume import *

def cubo_volume(lado):
    return (lado*lado*lado)

class TestCubo(unittest.TestCase):
    def test_cubo_volume(self):
        self.assertAlmostEqual(cubo_volume(2), 8)
        self.assertAlmostEqual(cubo_volume(1), 1)
        self.assertAlmostEqual(cubo_volume(0), 0)
        self.assertAlmostEqual(cubo_volume(5.5), 166.375)

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
````

O teste foi executado com sucesso, o que significa que a função `cubo_volume()` funcionou como se esperava.

### <span style="color: orange;">Pacote Numpy</span>

#### <span style="color: cyan;">Função ODEINT</span>

Usada na solução de sistemas de EDO's.  
Ela requer três parâmetros:

`y = odeint(modelo, y0, t)`  

- **modelo**: nome da função que retorna valores derivados para os valores 'y' e 't' solicitados, como `dydt = modelo(y, t)`
- **y0**: condições iniciais dos estados diferenciais
- **t**: pontos do tempo em que a solução deve ser reportada. Pontos internos adicionais são frequentemente calculados para manter a precisão da solução, mas não são reportados.

**Exemplo 1**:  
Qual é a solução da equação diferencial $dy(t)/dt = -k.y(t)$ com parâmetro $k=0.3$ e condição inicial $y0=5$.

````python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,t):          # funcao que retorna dy/dt
    k = 0.3
    dydt = -k * y
    return dydt
    
y0 = 5                   # condicao inicial
t = np.linspace(0,20)    # pontos no tempo
y = odeint(model,y0,t)   # solucao da EDO

plt.figure(figsize=(4,3)); plt.plot(t,y)
plt.xlabel('tempo'); plt.ylabel('y(t)'); plt.show()
````

**Exemplo 2**:  
Qual é a solução da equação diferencial $5.dy(t)/dt = -y(t)+2.u(t-10)$ com condição inicial $y(0)=1$.

````python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def u(t):                # sinal degrau unitario
    return 1*(t>=0)
    
def model(y,t):          # funcao que retorna dy/dt
    dydt = (-y + 2*u(t-10))/5.0
    return dydt
    
y0 = 1                    # condicao inicial
t = np.linspace(0,40,500) # pontos no tempo
y = odeint(model,y0,t)    # solucao da EDO

plt.figure(figsize=(4, 3)); plt.plot(t, y, 'r', label="y(t)"); plt.plot(t, 2*u(t-10), 'b', label="2u(t-10)")
plt.xlabel('tempo'); plt.ylabel('Amplitude'); plt.show()
````
